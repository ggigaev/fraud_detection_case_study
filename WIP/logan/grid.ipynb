```python
import numpy as np 
import pandas as pd 
from src import scrubbington
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import StratifiedShuffleSplit,GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_json('data.json')
```

```python
df.shape
```

```python
df['target'] = scrubbington.scrub_fraud_no_fraud(df)
```

```python
strat = StratifiedShuffleSplit(n_splits=3,test_size=0.2)
```

```python
X = df.drop('target',axis=1)
y = df.target
```

```python
for train_index,test_index in strat.split(X,y):
    x_train,x_test = X.loc[train_index],X.loc[test_index]
    y_train,y_test = y.loc[train_index],y.loc[test_index]
```

```python
gb = GradientBoostingClassifier(subsample=0.8,max_features=0.8)
```

```python
param_grid = {
    'max_depth':[4,7],
    'learning_rate': [0.05,0.01]
}
```

```python
grid = GridSearchCV(gb,param_grid,scoring='precision',n_jobs=-1,cv=strat)
```

```python
grid.fit(x_train,y_train)
```

```python
model = grid.best_estimator_
```

```python
model.fit(x_train,y_train)
```

```python
grid_y_pred = model.predict(x_test)
```

```python
accuracy_score(y_test,grid_y_pred)
```

```python
confusion_matrix(y_test,grid_y_pred)
```
